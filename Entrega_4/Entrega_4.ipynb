### Simulación de un ADC (cuantización y alias)

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 23 09:26:57 2025

@author: Ignacio
"""
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%%  Valores de parámetros
Fo = 1.0
Fs = 1000.0 # frecuencia de muestreo (Hz)
N = 1000   # cantidad de muestras
ts = 1/Fs # tiempo de muestreo
df = Fs/N # resolución espectral
Ac =2**0.5 #Amplitud 
DC = 0 #Valor Contínua
tita = 0 #Defasaje
SNR = 20 # SNR en dB
sigma = (10**(-SNR/10)) #Varianza
desvio = sigma**0.5 #Desvío Estándar
B = 4 #Número de Bits
fc = 0.8 #Factor de Carga para no tener saturación del ADC por ruido en el pico de la señal
Vref = Ac/fc #Tensión de referencia del ADC
#q = Vref/(2**(B-1)) 
```

```python
#%% Funciones
def my_sin_gen( vmax, dc, fo, ph, nn, fs): 
    tt = np.arange(0,nn*1/fs,1/fs) # grilla de sampleo temporal
    xx = vmax*np.sin(2*np.pi*fo*tt + ph) + dc #Senoidal
    return [tt, xx]

def cuantificar(signal, bits, vref):
    q = vref/(2**(bits-1)) 
    #s1 = (fc*Vref/Ac)*signal
    sq = signal*(1/q)
    sq1 = np.round(sq)
    ss_cuantizada = sq1*q#*(1/(fc*Vref/Ac))
    return ss_cuantizada
```

```python
#%%Obtengo señal con ruido y su versión cuantizada 

tt, s = my_sin_gen(vmax = Ac, dc = DC, fo = Fo, ph=tita, nn = N, fs = Fs )
#Genero señal aleatoria para el ruido
na=np.random.normal(0, desvio, len(tt))

ss = s + na

ssq = cuantificar(signal = ss,bits = B,vref = Vref)#4Bits
ssq2 = cuantificar(signal = ss,bits = B*2,vref = Vref)#8Bits

#Ploteo Señal Original y dos versiones Cuantizadas para 2 valores de Bits

plt.figure("Señal original vs Cuantizada",figsize=(10, 5))
plt.plot(tt, ss, label="Señal analógica")
plt.title("Señal analógica con Ruido Analógico")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.show()

plt.figure("Señal Cuantizada 1",figsize=(10, 5))
plt.plot(tt, ssq, color = 'r', label=f"Señal cuantizada con B = {B}", marker=".")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.show()

plt.figure("Señal Cuantizada 2",figsize=(10, 5))
plt.plot(tt, ssq2, color = 'g',label=f"Señal cuantizada con B = {2*B}", marker=".")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.show()
```

```python
Xq1 = ssq - ss#Estimación ruido de cuantización
Xq2 = ssq2 - ss#Estimación ruido de cuantización

plt.figure("Histograma")
plt.title(f"Histograma de la Diferencia entre señal Cuantizada y sin Cuantizar para {B} Bits")
plt.hist(ssq-ss)
plt.show()

plt.figure("Histograma")
plt.title(f"Histograma de la Diferencia entre señal Cuantizada y sin Cuantizar para {2*B} Bits")
plt.hist(ssq2-ss)
plt.show()

q1 = Vref/(2**(B-1))/2
q2 = Vref/(2**(2*B-1))/2

print('q1=',q1)#q para 4 Bits
print('q2=',q2)#q para 8 Bits
```

```python
#%%Espectro de señal cuantizada y de la estimación de ruido

TFF=np.fft.fft(ss)/N#Espectro de la cuantizada con 8 bits
mod=10*np.log10(2*np.abs(TFF)**2)
TFF1=np.fft.fft(ssq)/N#Espectro de la señal
mod1=10*np.log10(2*np.abs(TFF1)**2)
TFF2=np.fft.fft(ssq2)/N#Espectro de la señal
mod2=10*np.log10(2*np.abs(TFF2)**2)

plt.figure("DFT Senoidal Cuantizada vs Sin Cuantizar")
plt.title("DFT Senoidal Sin Cuantizar")
plt.plot(tt*len(tt)*df, mod,color = 'b',label="Espectro Señal sin cuantizar")
plt.xlabel("Freq (Hz)")
plt.ylabel("DFT Amplitude |X(freq)|")
plt.legend()
plt.tight_layout()
plt.show()
plt.figure("DFT Senoidal Cuantizada 1")
plt.plot(tt*len(tt)*df, mod1,color = 'r',label=f"Espectro Señal cuantizada con B = {B}")
plt.xlabel("Freq (Hz)")
plt.ylabel("DFT Amplitude |X(freq)|")
plt.legend()
plt.show()
plt.figure("DFT Senoidal Cuantizada 2")
plt.plot(tt*len(tt)*df, mod2,color = 'g',label=f"Espectro Señal cuantizada con B = {2*B}")
plt.xlabel("Freq (Hz)")
plt.ylabel("DFT Amplitude |X(freq)|")
plt.legend()
plt.show()
```

```python
TFFean=np.fft.fft(na)/N#Espectro de la estimacion de ruido de cuantizacion
modea=10*np.log10(2*np.abs(TFFean)**2)
TFFq1=np.fft.fft(Xq1)/N#Espectro del ruido 4 Bits
modq1=10*np.log10(2*np.abs(TFFq1)**2)#10*np.log10(escala)
TFFq2=np.fft.fft(Xq2)/N#Espectro del ruido 8 Bits
modq2=10*np.log10(2*np.abs(TFFq2)**2)#10*np.log10(escala)

plt.figure("DFT Ruido Analógico")

plt.title('DFT Ruido Analógico')
plt.plot(tt*len(tt)*df, modea,color = 'b',label="Estimacion de ruido analógico")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.legend()
plt.figure("DFT Ruido de Cuantización B=4")
plt.title(f'DFT Ruido de Cuantización B={B}')
plt.plot(tt*len(tt)*df, modq1,color = 'r',label=f"DFT Error de Cuantización para B={B}")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.legend()
plt.tight_layout()
plt.show()
plt.figure("DFT Ruido de Cuantización B=8")
plt.title(f'DFT Ruido de Cuantización B={2*B}')
plt.plot(tt*len(tt)*df, modq2,color = 'g',label=f"DFT Error de Cuantización para B={2*B}")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
TFFssq=np.fft.fft(ssq)/N#Espectro de la cuantizada con 8 bits
modssq=10*np.log10(2*np.abs(TFFssq)**2)

plt.figure("DFT Senoidal Cuantizada + Xq + Ruido")
plt.title(f"DFT Senoidal Cuantizada + Xq + Ruido para {B} Bits con SNR={SNR}dB")
plt.plot(tt*len(tt)*df, modssq,color = 'b',label=f"Espectro Señal cuantizada con B = {B}")
plt.plot(tt*len(tt)*df, modq1,color = 'r',label="Estimacion de ruido Xq - Xa")
plt.plot(tt*len(tt)*df, modea,color = 'g',label=f'Ruido con $\\sigma^2$ = {sigma:3.3}')
plt.legend()
plt.show()

TFFssq2=np.fft.fft(ssq2)/N#Espectro de la cuantizada con 8 bits
modssq2=10*np.log10(2*np.abs(TFFssq2)**2)

plt.figure("DFT Senoidal Cuantizada + Xq + Ruido")
plt.title(f"DFT Senoidal Cuantizada + Xq + Ruido para {2*B} Bits con SNR={SNR}dB")
plt.plot(tt*len(tt)*df, modssq2,color = 'b',label=f"Espectro Señal cuantizada con B = {2*B}")
plt.plot(tt*len(tt)*df, modq2,color = 'r',label="Estimacion de ruido Xq - Xa")
plt.plot(tt*len(tt)*df, modea,color = 'g',label=f'Ruido con $\\sigma^2$ = {sigma:3.3}')
plt.legend()
plt.show()
```

# Conclusiones

Se simularon y graficaron distintos casos, donde se varió la SNR de la señal analógica y también la cantidad de bits del ADC. Esto generó resultados muy interesantes, por un lado se observa que si la SNR es muy alta y se tiene una señal muy pura, se deberá utilizar un ADC de mejor calidad ya que en caso contrario el ruido de cuantización estará muchos decibeles por encima del ruido analógico e interferirá con el análisis de la señal. Por otro lado, si la SNR es muy baja y se tiene una señal muy ruidosa, es un desperdicio utilizar tantos bit del ADC ya que se estaría haciendo una cuantización extremadamente exigente cuando el ruido analógico predominará fácilmente. Por lo tanto se concluye que la mejor respuesta a este problema sería siempre utilizar una cuantización que genere un ruido en el orden del ruido de la señal analógica para que no se desperdicien recursos ni tampoco para que se agregue más ruido a la señal.
Según el Holton se llega a una expresión que relaciona el ruido analógico con el ruido de cuantización, la cual dependerá de la tensión de referencia del ADC y la cantidad de bits del mismo, de forma que:
