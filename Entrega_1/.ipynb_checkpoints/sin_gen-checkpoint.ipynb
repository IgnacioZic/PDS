### Generador de una señal seno con parámetros

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%%  Valores de parámetros
Fo = 1.0
Fs = 1000.0 # frecuencia de muestreo (Hz)
N = 1000   # cantidad de muestras
ts = 1/Fs # tiempo de muestreo
df = Fs/N # resolución espectral
Ac =1 #Amplitud 
DC = 0 #Valor Contínua
tita= 0 #Defasaje
```

```python
#%% Funcion
def my_sin_gen( vmax, dc, fo, ph, nn, fs): 
    tt = np.arange(0,(nn-1)*1/fs,1/fs) # grilla de sampleo temporal
    xx = vmax*np.sin(2*np.pi*fo*tt + ph) + dc #Senoidal
    return [tt, xx]
#%%
```

```python
#%% Llamada a la función
tt, s = my_sin_gen(vmax = Ac, dc = DC, fo = Fo, ph=tita, nn = N, fs = Fs )
#s = Ac*np.sin(2*np.pi*fo*tt + tita) + DC #Senoidal


plt.figure(1)
plt.title('Senoidal')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud')
plt.plot(tt, s)
plt.show()
```

```python
tt, s = my_sin_gen(vmax = 5, dc = 1, fo = Fo, ph= np.pi/2, nn = N, fs = Fs )
#s = Ac*np.sin(2*np.pi*fo*tt + tita) + DC #Senoidal


plt.figure(1)
plt.axis([0, 1, -5, 7])
plt.title('Cosenoidal')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud')
plt.plot(tt, s)
plt.show()
```

```python
tt, s = my_sin_gen(vmax = 3, dc = 2, fo = 2*Fo, ph= np.pi/2, nn = 4*N, fs = 4*Fs )
#s = Ac*np.sin(2*np.pi*fo*tt + tita) + DC #Senoidal


plt.figure(1)
plt.axis([0, 1, -2, 6])
plt.title('Cosenoidal del doble de Fo')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud')
plt.plot(tt, s)
plt.show()
```

```python
tt, s = my_sin_gen(vmax = Ac, dc = DC, fo = 4*Fo, ph= tita, nn = N, fs = 2*Fs )
#s = Ac*np.sin(2*np.pi*fo*tt + tita) + DC #Senoidal


plt.figure(1)
plt.axis([0, 0.6, -2, 2])
plt.title('Senoidal')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud')
plt.plot(tt, s)
plt.show()
```

```python
tt, s = my_sin_gen(vmax = Ac, dc = DC, fo = 4*Fo, ph= tita, nn = 2*N, fs = Fs )
#s = Ac*np.sin(2*np.pi*fo*tt + tita) + DC #Senoidal


plt.figure(1)
plt.axis([0, 1, -2, 2])
plt.title('Senoidal de 4 Fo')
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud')
plt.plot(tt, s)
plt.show()
```

## Conclusiones
Se logró con éxito generar la senoidal caracterizada con los parámetros deseados.
Se observa que al defasar pi/2 la senoidal se convierte en un coseno. Si se aumenta mucho la Fs y no la N se modifica la Resolución Espectral y se observa sólo un período o más de uno según cambie la relación. Si se modifica la Fo entrarán más períodos en nuestra ventana de observación. Si no se cumple Nyquist se rompe todo y no se puede ver la senoidal.
